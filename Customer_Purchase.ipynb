```
#separarting ordinal categorical variable
df1=df.iloc[:,2:4]
df1.head(2)
```

```
#implementing ordinal encoding on df1
from sklearn.preprocessing import OrdinalEncoder

```

```
oe=OrdinalEncoder(categories=[['School','UG','PG'],['Poor','Average','Good']])
```

```
df1_oe=oe.fit_transform(df1)
df1_oe
```

```
print(type(df1_oe))
```

```
df1=pd.DataFrame(df1_oe,columns=['Education','Review'])
df1.head()
```

```
df1=df1.astype(int)
df1.head(2)
```

```
#onehotencoder
from sklearn.preprocessing import OneHotEncoder
```

```
oh_enc=OneHotEncoder(sparse_output=False)
#it provide output in array format
```

```
oh_enc_arr=oh_enc.fit_transform(df2)
oh_enc_arr
```

```
oh_enc=OneHotEncoder(sparse_output=False, drop='first')
#it provide output in array format
```

```
oh_enc_arr=oh_enc.fit_transform(df2)
oh_enc_arr
```

```
df2=pd.DataFrame(oh_enc_arr,columns=['Gender','Purchased'])
df2.head()
```
